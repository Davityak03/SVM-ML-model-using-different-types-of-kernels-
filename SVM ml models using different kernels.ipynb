```python
#Data Pre-processing Step  
# importing libraries  
import numpy as nm  
import matplotlib.pyplot as mtp  
import pandas as pd  
from sklearn.metrics import accuracy_score
import seaborn as sns  
    
#importing datasets  
data_set= pd.read_csv('banana_dataset.csv')  
data_set
```

```python
#Extracting Independent and dependent Variable  
x= data_set.iloc[:, [0,1]].values  
y= data_set.iloc[:, 2].values  
 
```

```python
x
```

```python
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0)  
#feature Scaling  
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test)  
```

```python
list1=[]
from sklearn.svm import SVC # "Support vector classifier"  
#taking kernel as rbf
classifier = SVC(kernel='rbf', random_state=0)  
classifier.fit(x_train, y_train)
#Predicting the test set result  
y_pred= classifier.predict(x_test)  
#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot=True,fmt=".2f",cmap="Blues")
ac=accuracy_score(y_test, y_pred)
print("Accuracy is:",ac)
list1.append(ac)
```

```python
#taking kernel as linear
classifier1=SVC(kernel='linear', random_state=0)  
classifier1.fit(x_train, y_train)
#Predicting the test set result  
y_pred=classifier1.predict(x_test)  
#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix  
cm1=confusion_matrix(y_test, y_pred)
ac1=accuracy_score(y_test, y_pred)
print("Acuracy is:",ac1)
list1.append(ac1)
sns.heatmap(cm1,annot=True,fmt=".2f",cmap="Blues")
```

```python
#taking kernel as poly
classifier2=SVC(kernel='poly', random_state=0)  
classifier2.fit(x_train, y_train)
#Predicting the test set result  
y_pred=classifier2.predict(x_test)  
#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix  
cm2=confusion_matrix(y_test, y_pred)
sns.heatmap(cm2,annot=True,fmt=".2f",cmap="Blues")
ac2=accuracy_score(y_test, y_pred)
print("Accuracy is:",ac2)
list1.append(ac2)
```

```python
#taking kernel as sigmoid
classifier3=SVC(kernel='sigmoid', random_state=0)  
classifier3.fit(x_train, y_train)
#Predicting the test set result  
y_pred=classifier3.predict(x_test)  
#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix  
cm3=confusion_matrix(y_test, y_pred)
sns.heatmap(cm3,annot=True,fmt=".2f",cmap="Blues")
ac3=accuracy_score(y_test, y_pred)
print("Accuracy is:",ac3)
list1.append(ac3)
```

```python
#comparitive analysis between their accuracy
list2=['RBF','LINEAR','POLY','SIGMOID']
import matplotlib.pyplot as plt
plt.bar(list2,list1,width=0.25)
plt.xlabel("Kernels")
plt.ylabel("Accuracy")
plt.title("SVM")
plt.show()
```
